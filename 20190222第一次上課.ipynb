```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# 六芒星
x = (0,6,3,0)
y = (1.732050807568877,1.732050807568877,6.928203230275509,1.732050807568877)
plt.plot(x, y,color = 'black')

x1 = (0,3,6,0)
y1 = (5.196152422706632,0,5.196152422706632,5.196152422706632)
plt.plot(x1, y1,color = 'black')

# 內圓
r = 3.464101615137755
a, b = (3, 3.464101615137755)
theta = np.arange(0, 2*np.pi, 0.01)
x2 = a + r * np.cos(theta)
y2 = b + r * np.sin(theta)
plt.plot(x2, y2,color = 'black')

# 外圓
r = 4
a, b = (3, 3.464101615137755)
theta = np.arange(0, 2*np.pi, 0.01)
x2 = a + r * np.cos(theta)
y2 = b + r * np.sin(theta)
plt.plot(x2, y2,color = 'black')

#小圈圈
r = 0.2679491924311227
a, b = (3, 3.464101615137755)
theta = np.arange(0, 2*np.pi, 0.01)

for i in range(0,45):
    x2 = a + ( 3.464101615137755 + 0.2679491924311227) *np.cos(i) + r * np.cos(theta)
    y2 = b +  ( 3.464101615137755 + 0.2679491924311227)*np.sin(i) + r * np.sin(theta)
    plt.plot(x2, y2,color = 'black')
```
